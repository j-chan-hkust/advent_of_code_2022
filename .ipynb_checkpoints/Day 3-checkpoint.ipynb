```python
rucksacks = input_str.split("\n")
```

```python
count = 0
for rucksack in rucksacks:
    A = rucksack[:int(len(rucksack)/2)]
    B = rucksack[int(len(rucksack)/2):]
    chars = set()
    for char in A:
        chars.add(char)
    for char in B:
        if char in chars:
            if char.islower():
                count += ord(char)-96
            else:
                count += ord(char)-38
            break
```

```python
counter = 0
count = 0
while counter < len(rucksacks):
    elf_group = rucksacks[counter:counter+3]
    elf_group = [set(rucksack) for rucksack in elf_group]
    
    #in my original implementation, I mistakenly used dictionarys and manually iterate through, intead of making use of set logic to make my life easier
    #you can see vestiges of this implementation in the code for solving part 1 
    badge = elf_group[0].intersection(elf_group[1],elf_group[2]).pop()
    
    if badge.islower():
        count += ord(badge)-96
    else:
        count += ord(badge)-38
    counter += 3
    #print(badge)
```
