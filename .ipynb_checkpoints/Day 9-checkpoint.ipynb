```python
input_str = """R 4
U 4
L 3
D 1
R 4
D 1
L 5
R 2"""
```

```python
instructions = input_str.split('\n')
instructions
```

```python
def normalize_1d(number):
    if abs(number) == 1:
        return number
    else:
        return 1 if number>0 else -1
```

```python
def squared_euclidean_dist(displacement_x, displacement_y):
    return int((displacement_x)**2 + (displacement_y)**2)
```

```python
head_location = (0,0)
tail_location = (0,0)
tail_visited = {(0,0)}
direction_dict = {'U':(0,1), 'D':(0,-1) , 'R':(1,0), 'L':(-1,0)} #to be honest, order is arbitrary, they just need to all be diff
for instruction in instructions:
    direction, distance = instruction.split(' ')
    distance = int(distance)
    direction = direction_dict[direction]
    
    while distance>0:
        distance -= 1
        head_location = tuple(map(lambda i, j: i + j, head_location, direction))
        #then we move the tail
        x_displacement = head_location[0]-tail_location[0]
        y_displacement = head_location[1]-tail_location[1]
        if squared_euclidean_dist(x_displacement,y_displacement)<= 2:
            #do nothing
            pass
        elif x_displacement == 0 or y_displacement == 0:
            #just move 1 step u d l r
            if x_displacement != 0:
                tail_location = (tail_location[0] + normalize_1d(x_displacement), tail_location[1])
            else:
                tail_location = (tail_location[0], tail_location[1] + normalize_1d(y_displacement))
        else:
            tail_location = (tail_location[0] + normalize_1d(x_displacement), tail_location[1] + normalize_1d(y_displacement))
        tail_visited.add(tail_location)
```

```python
def move_tail_knot(head_location, tail_location): # head and tail is a pair of tuples
    #we now update the tail location
    x_displacement = head_location[0]-tail_location[0]
    y_displacement = head_location[1]-tail_location[1]
    if squared_euclidean_dist(x_displacement,y_displacement)<= 2:
        #do nothing
        pass
    elif x_displacement == 0 or y_displacement == 0:
        #just move 1 step u d l r
        if x_displacement != 0:
            tail_location = (tail_location[0] + normalize_1d(x_displacement), tail_location[1])
        else:
            tail_location = (tail_location[0], tail_location[1] + normalize_1d(y_displacement))
    else:
        tail_location = (tail_location[0] + normalize_1d(x_displacement), tail_location[1] + normalize_1d(y_displacement))
    
    return tail_location
```

```python
knots = [(0,0) for i in range(10)] #each knot is its own object
tail_visited = {(0,0)}
direction_dict = {'U':(0,1), 'D':(0,-1) , 'R':(1,0), 'L':(-1,0)} #to be honest, order is arbitrary, they just need to all be diff

for instruction in instructions:
    direction, distance = instruction.split(' ')
    distance = int(distance)
    direction = direction_dict[direction]
    
    while distance>0:
        distance -= 1
        knots[0] = tuple(map(lambda i, j: i + j, knots[0], direction))
        for i in range(len(knots)-1):
            knots[i+1] = move_tail_knot(knots[i],knots[i+1]) #knots 0 is head, knots 9 is tail
        tail_visited.add(knots[9])
        print(knots)
```
