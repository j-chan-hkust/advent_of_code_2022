```python
input_str = """    [D]    
[N] [C]    
[Z] [M] [P]
 1   2   3 

move 1 from 2 to 1
move 3 from 1 to 3
move 2 from 2 to 1
move 1 from 1 to 2"""
```

```python
crates, instructions = input_str.split('\n\n')
```

```python
crates = crates.split('\n')
```

```python
transposed_crates = [''.join([row[i] for row in crates]) for i in range(len(crates[0]))]    
```

```python
transposed_crates = [transposed_crates[i].strip()[:-1] for i in range(1, len(transposed_crates),4)]
```

```python
crates = transposed_crates # note that 0th item of any string is the top of the stack.
crates
```

```python
instructions = instructions.split('\n')
```

```python
import re

for instruction in instructions:
    _, move, start, end =re.split(r'\D+',instruction)
    move = int(move)
    start = int(start)-1 # keep forgetting that advent likes indexing from 1...
    end = int(end)-1
    
    crates[end] = crates[start][0:move][::-1] + crates[end]
    crates[start] = crates[start][move:]
    
    print(crates)
    
```

```python
solution = ''.join([stack[0] for stack in crates])
```
