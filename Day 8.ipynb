```python
tree_array = [[int(char) for char in row] for row in input_str.split('\n')]
```

```python
# while we could iteratively look at every tree and check if it is visible, it makes more sense to just check from each of the directions
# for n x n grid, are 4n directions, and n^2 trees
# with a naive approach where you check each tree, your complexity could be n^3
# just checking each direction and checking the visibility, your complexity is only n^2
```

```python
visible_trees = set() # a set of coordinates of the visible trees

# look left:
for i in range(len(tree_array)):
    max_height = -1
    for j in range(len(tree_array[0])):
        if tree_array[i][j]>max_height:
            max_height = tree_array[i][j] #cant be bothered to optimize once we've seen the "highest tree"
            visible_trees.add((i, j))

# look right:
for i in range(len(tree_array)):
    max_height = -1
    for j in reversed(range(len(tree_array[0]))):
        if tree_array[i][j]>max_height:
            max_height = tree_array[i][j] #cant be bothered to optimize once we've seen the "highest tree"
            visible_trees.add((i, j))

# look down:
for j in range(len(tree_array[0])):
    max_height = -1
    for i in range(len(tree_array)):
        if tree_array[i][j]>max_height:
            max_height = tree_array[i][j] #cant be bothered to optimize once we've seen the "highest tree"
            visible_trees.add((i,j))

# look up:
for j in range(len(tree_array[0])):
    max_height = -1
    for i in reversed(range(len(tree_array))):
        if tree_array[i][j]>max_height:
            max_height = tree_array[i][j] #cant be bothered to optimize once we've seen the "highest tree"
            visible_trees.add((i,j))
```

```python
# unfortunately, part 2 requires something more inelegant (or I just can't think of a more efficient solution)
# for each tree (n^2) we need to check each of the 4 cardinal directions (2n). complexity ends up as n^3
```

```python
max_score = 0
for i in range(len(tree_array)):
    for j in range(len(tree_array[0])):
        # we calculate the scenic score for each tree...
        current_tree_height = tree_array[i][j]
        count = 0
        score = 1
        for x in reversed(range(i)): # i to 0
            count += 1
            if current_tree_height<=tree_array[x][j]:
                break
        score *= count
        count = 0
        for x in range(i+1,len(tree_array)): # i to the end (98)
            count += 1
            if current_tree_height<=tree_array[x][j]:
                break
        score *= count
        count = 0
        for y in reversed(range(j)): # j to 0
            count += 1
            if current_tree_height<=tree_array[i][y]:
                break
        score *= count
        count = 0
        for y in range(j+1,len(tree_array[0])): # j to the end (98)
            count += 1
            if current_tree_height<=tree_array[i][y]:
                break
        score *= count
        
#         print(score)
#         print(i,j)
#         print(current_tree_height)
        if score>max_score:
            max_score = score
            
```
